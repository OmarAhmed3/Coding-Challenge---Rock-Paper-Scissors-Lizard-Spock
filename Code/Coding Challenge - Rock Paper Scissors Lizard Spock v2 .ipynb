### 1. Import Libraries

```python
###################################################### import libraries###########################################
import pandas, numpy
import random
from enum import IntEnum
```

### 2. make a Class of Actions

```python
###################################### make a class of Actions#################################################
"""
Using enum.IntEnum allows you to create attributes and assign them values similar to those shown above.
This helps clean up your code by grouping actions into their own namespaces and making the code more expressive.
"""
class Action(IntEnum):
    Rock = 0
    Paper = 1
    Scissors = 2
    Lizard = 3
    Spock = 4
```

### 3. User Function  

```python
###################################### User Function ###################################

"""
Using get_user_selection() fuuction to user enters a value from a class of actions and return it this value.
make a Exception if user enters a value out of a class of Actions and help him to enter a correct value.
"""
def get_user_selection():
	choices = [f"{action.name}[{action.value}]" for action in Action]
	choices_str = ", ".join(choices)
	selection = int(input(f"Enter a choice ({choices_str}): "))
	action = Action(selection)
	return action
```

```python
while True:
	try:
		user_action = get_user_selection()
		print(user_action)
		break
	except ValueError as e:
		range_str = f"[0, {len(Action) - 1}]"
		print(f"Invalid selection. Enter a value in range {range_str}")
```

### 4. Computer Function 

```python
###################################### Computer Function #################################################

"""
Using get_computer_selection() fuuction to computer enters a value from a class of actions and return it this value.
"""
def get_computer_selection():
    selection = random.randint(0, len(Action) - 1)
    action = Action(selection)
    return action
```

```python
computer_action = get_computer_selection()
print(computer_action)
```

### 5. Game Function 

```python
####################################### Game Function #######################################
"""""
 you can use a dictionary to help show the relationships between actions.
 
 Dictionaries are a great way to show a key-value relationship. 
 
 In this case, the key can be an action, like scissors, and the value can be a list of actions that it beats.

 while condition if you to play again press a keyboard y.


"""""
def determine_winner(user_action, computer_action):
	while True:

		victories = {
		Action.Scissors: [Action.Lizard, Action.Paper],
		Action.Paper: [Action.Spock, Action.Rock],
		Action.Rock: [Action.Lizard, Action.Scissors],
		Action.Lizard: [Action.Spock, Action.Paper],
		Action.Spock: [Action.Scissors, Action.Rock]
         }

		defeats = victories[user_action]
		if user_action == computer_action:
			print(f"Both players selected {user_action.name}. It's a tie!")
		elif computer_action in defeats:
			print(f"{user_action.name} beats {computer_action.name}! You win!")
		else:
			print(f"{computer_action.name} beats {user_action.name}! You lose.")
		play_again = input("Play again? (y/n): ")
		if play_again.lower() != "y":
			break
		else:
			try:
				user_action = get_user_selection()
				print(user_action.name)
				computer_action = get_computer_selection()
				print(computer_action.name)
			except ValueError as e:
				range_str = f"[0, {len(Action) - 1}]"
				print(f"Invalid selection. Enter a value in range {range_str}")
```

```python
determine_winner(user_action,computer_action)
```
